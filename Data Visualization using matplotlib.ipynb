```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#1:

A = [1,2,3,4]
B = [1,3,4,2]
```

```python
#2:

plt.figure(figsize = (8,4))#allow you to manipulate figure size
plt.plot(A,B,
         linestyle = "--",#"-"stands for a solid line
        marker = "o",
        markersize = 12,
        color = "b",#b stands for blue
        linewidth = 3)
plt.ylim(0,5)#Good to start at 0; Same for x
plt.xlabel("X Label",fontsize = 12)
plt.ylabel("Y Label",fontsize = 12)
plt.title("I am a title",fontsize = 14, fontweight = "bold")
plt.show()
```

```python
C = [0,1,2,3]
D = [0.5,2,1,4]
plt.figure(figsize = (8,4))
plt.plot(A,B,
         linestyle = "--",
        marker = "o",
        markersize = 12,
        color = "b",
        linewidth = 3)

plt.plot(C,D,
        linestyle = ":",
        marker = "s",
        markersize = 8,
        color = "r",
        linewidth = 1)
plt.ylim(0,5)
plt.xlabel("X Label",fontsize = 12)
plt.ylabel("Y Label",fontsize = 12)
plt.title("This is my first Python plot",fontsize = 14, fontweight = "bold")
plt.show()
```

```python
#6.
np.random.seed(10)#
iq_scores = np.random.normal(100,15,100)#np.random.normal(mean,std,size)
iq_scores.mean()
```

```python
plt.hist(iq_scores)
plt.axvline(x = iq_scores.mean(), color = "red")
plt.axvline(x = iq_scores.mean() + iq_scores.std(), color = "red" , 
            linestyle = "--")
plt.show()
```
